```python
#Banking Transaction Simulation
import random

#Declaring type of transaction, and amount of money that will be randomly generated
transaction_types = ['Deposit', 'Withdraw', 'Transfer'] #setting the three types of transcations that we are going to use
min_amount = 10 
max_amount = 1000 
num_transactions = 5000 

#Randomly assigning transaction type and amount to elements and storing them in transactions
transactions = [] 
for i in range(num_transactions):
    transaction_type = random.choice(transaction_types)
    amount = random.randint(min_amount, max_amount)  
    transactions.append(f"Transaction {i + 1}: {transaction_type} ${amount}") 

#Simulating the data_nodes
data_nodes = {
    "DataNode 1 ": [],
    "DataNode 2 ": [],
    "DataNode 3 ": [], 
}

#Appending the transactions in data nodes
for transaction in transactions:
    node = random.choice(list(data_nodes.keys()))
    data_nodes[node].append(transaction)

#Printing Distributed data
print("Distributed data: ")
for node, transaction in data_nodes.items():
    print(f"\n{node}: {transactions[:10]}")


#Implementing a replication mechanism where each transaction has 50% chance of being replicated to a random node, and no more copies of each transaction
#Exist
replication_factor = 2
for node in data_nodes:
    for transaction in data_nodes[node]:
        if random.random() < 0.5:
            target_node = random.choice([n for n in data_nodes if n != node])
            data_nodes[target_node].append(transaction)

#Printing data after replication
print("\nData after replication: ")
for node, transaction in data_nodes.items():
    print(f"\n{node} : {transaction[:10]}")

#Defining function to retrieve specific transaction if found in node
def retrieve_transaction(search_transaction):
    for node, transaction_list in data_nodes.items():
        if search_transaction in transaction_list:
            return f"\nFound in {node}: {search_transaction}"
    return f"\n{search_transaction} not found in any DataNode"
```

```python
#E-commerce order simulation
import random

#Setting the conditions for our orders
num_orders = 10000
customer_orders = 1000
min_amount = 5
max_amount = 500

#Assigning random values to each of the three variables that we generated
orders = []
for i in range(num_orders):
    order_id = f"OrderID: {i + 1}"
    costumer_id = f"CostumerID: {random.randint(1, customer_orders)} "
    amount = random.randint(min_amount,max_amount)
    orders.append(f"{order_id} {costumer_id} Amount ${amount}")

#Declaring data_nodes
data_nodes = {
"DataNode A" : [],
"DataNode B" : [],
"DataNode C" : [],
}

#Assigning orders to nodes randomly
for order in orders:
    node = random.choice(list(data_nodes.keys()))
    data_nodes[node].append(order)


#Printing Distributed data
print("Distributed data: ")
for node, order_list in data_nodes.items():
    print(f"\n{node}: {order_list[:10]}")

#Randomly assigning replicated data to nodes
replication_factor = 2
for node in data_nodes:
    for order in data_nodes[node]:
        if random.random() < 0.5:
            target_node = random.choice([n for n in data_nodes if n != node])
            data_nodes[target_node].append(order)


#Printing Replicated data
print("\nData after replication: ")
for node, order_list in data_nodes.items():
    print(f"\n{node} : {order_list[:10]}")

#function to retrieve order
def retrieve_order(search_order):
    for node, order_list in data_nodes.items():
        for order in order_list:  
            if search_order in order:  
                return f"\nFound in {node}: {order}"
    return f"\n{search_order} not found in any DataNode"

#Testing retrieve_order function
print(retrieve_order("OrderID: 2"))
print(retrieve_order("OrderID: 3"))
print(retrieve_order("OrderID: 999999")) #does not exist
```
